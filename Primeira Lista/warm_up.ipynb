```python
'''Warm up 7: Crie classes para representar estados e cidades. Cada estado tem um nome, sigla e 
cidades. Cada cidade tem nome e população. Escreva um programa de testes que crie 
três estados com algumas cidades em cada um. Exiba a população de cada estado 
como a soma da população de suas cidades.'''

class Estado():
    def __init__(self, nome: str, sigla: str, cidades: list, populacao_total = 0):
        self.nome = nome 
        self.sigla = sigla
        self.cidades = cidades
        self.populacao_total = populacao_total
        
    def soma_populacao(self):
        for cidade in self.cidades:
            self.populacao_total += cidade.populacao        

class Cidade():
    def __init__(self, nome: str, populacao: int):
        self.nome = nome
        self.populacao = populacao
        
def main():
    cidade1 = Cidade('Florianópolis', 500000)
    cidade2 = Cidade('Palhoça', 150000)
    cidade3 = Cidade('Chapecó', 200000)
    
    estado1 = Estado('Santa Catarina', 'SC', [cidade1, cidade2, cidade3])
    estado1.soma_populacao()
    
    print(f'O estado de {estado1.nome} possui {estado1.populacao_total} habitantes.')
    
    cidade4 = Cidade('Terra de Areia', 11300)
    cidade5 = Cidade('Caxias do Sul', 500000)
    cidade6 = Cidade('Capão da Canoa', 50000)
    
    estado2 = Estado('Rio Grande do Sul', 'RS', [cidade4, cidade5, cidade6])
    estado2.soma_populacao()
    
    print(f'O estado de {estado2.nome} possui {estado2.populacao_total} habitantes.')
    
    cidade7 = Cidade('Curitiba', 1900700)
    cidade8 = Cidade('Londrina', 580000)
    cidade9 = Cidade('Maringá', 400000)
    
    estado3 = Estado('Paraná', 'PR', [cidade7, cidade8, cidade9])
    estado3.soma_populacao()
    
    print(f'O estado de {estado3.nome} possui {estado3.populacao_total} habitantes.')
    
main()
```

```python
'''Warm up 8: Escreva uma classe Coordenada com atributos x e y, e métodos para mostrar as 
coordenadas, calcular a distancia para outra coordenada, comparar coordenadas, 
mostrar no formato coordenada polar.'''

import math

class Coordenada():
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y
        
    def mostrar_coordenadas(self):
        print(f'As coordenadas são ({self.x}, {self.y})')
        
    def calcular_distancia(self, other):
        distancia = math.sqrt((other.x - self.x)**2 + (other.y - self.y)**2)
        print('A distância é de %.2f' % distancia)
    
    def comparar_coordenadas(self, other):
        if self.x == other.x and self.y == other.y:
            print('Coordenadas identicas')
        else:
            print('Coordenadas diferentes!')
    
    def coordenada_polar(self):
        r = math.sqrt(self.x**2 + self.y**2)
        teta = math.atan(self.y/self.x)
        print('As coordenadas polares são (%.2f, %.2f)' % (r, teta))
        
def main():
    coordenadas1 = Coordenada(1, 2)
    coordenadas2 = Coordenada(1, 2)
    coordenadas1.mostrar_coordenadas()
    coordenadas1.coordenada_polar()
    coordenadas1.comparar_coordenadas(coordenadas2)
    coordenadas1.calcular_distancia(coordenadas2)
    
main()
```

```python
'''Warm up 9: Escreva classes para as seguintes formas: quadrado, retângulo e círculo.'''

class Quadrado:
    def __init__(self, lado: float):
        self.lado = lado
    def Area(self):
        área = self.lado**2
        print('A área do quadrado é %.2f um' % área)
        
class Retangulo:
    def __init__(self, base: float, altura: float):
        self.base = base
        self.altura = altura
    def Area(self):
        área = self.base*self.altura
        print('A área do retângulo é %.2f um' % área)
        
class Circulo:
    def __init__(self, raio: float):
        self.raio = raio
    def Area(self):
        área = self.raio**2*3.14
        print('A área do circulo é %.2f um' % área)
        
def main():
    quadrado1 = Quadrado(2)
    retangulo1 = Retangulo(2, 2)
    circulo1 = Circulo(2)
    
    quadrado1.Area()
    retangulo1.Area()
    circulo1.Area()

main()
```

```python
'''Warm up 10: Escreva uma classe Fracao que armazena dois inteiros, numerador e denominador. 
a) Implemente metodos para somas, subtração, multiplicação e divisão de duas 
frações 
b) Implemente o método que imprime uma fração no formato numerador / 
denominador 
c) Implemente um método que inverte a fração 
d) Implemente um método que retorna a fração em valor real 
e) Implemente um método que cria uma fração (numerador/denominador) a partir 
de um número real'''

from fractions import Fraction

class Fracao():
    def __init__(self, numerador: int, denominador: int):
        self.numerador = numerador
        self.denominador = denominador
        
    def __str__(self):
        if self.numerador == 0:
            return "0"
        elif self.denominador == 1:
            return str(self.numerador)
        else:
            return str(self.numerador)+'/'+ str(self.denominador)

    def _add_(self, other):
        den = self.denominador * other.denominador
        num = self.numerador * other.denominador + self.denominador * other.numerador
        return (Fracao(num, den).simplifica())
    
    def _sub_(self, other):
        den = self.denominador * other.denominador
        num = self.numerador * other.denominador - self.denominador * other.numerador
        return (Fracao(num, den).simplifica())
    
    def _mul_(self, other):
        num = self.numerador * other.numerador
        den = self.denominador * other.denominador
        return (Fracao(num, den).simplifica())
        
    def divisao(self, other):
        num = self.numerador * other.denominador
        den = self.denominador * other.numerador
        return (Fracao(num, den).simplifica())
    
    def inverte_fracao(self):
        num = self.denominador
        den = self.numerador
        return (Fracao(num, den).simplifica())
        
    def valor_real(self):
        valor = self.numerador/self.denominador
        return valor
        
    def mdc(a, b):
        while b != 0:
            resto = a % b
            a = b
            b = resto
        return a

    def simplifica(self):
        mdc = Fracao.mdc(self.numerador, self.denominador)
        self.numerador = self.numerador//mdc
        self.denominador = self.denominador//mdc
        return self
    
    def tornar_fracao(num: float):
        return Fraction(num)
        
def main():
    fracao1 = Fracao(7, 2)  
    fracao2 = Fracao(4, 2)
    
    print(fracao1._add_(fracao2))
    print(fracao1._mul_(fracao2))
    print(fracao1._sub_(fracao2))
    print(fracao1.divisao(fracao2))
    print(fracao1.inverte_fracao())
    print(fracao1.valor_real())
    print(Fracao.tornar_fracao(0.25))

main()
```
